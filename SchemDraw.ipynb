### Mesh method

```python
with schemdraw.Drawing() as d:
    d.config(unit=5)
    V1 = elm.SourceV().up().label('V1')
    R1 = elm.Resistor().right().label('R1')
    d.push()
    elm.Resistor().right().label('R2')
    V2=elm.SourceV().down().reverse().label('V2')
    L3=elm.Line().left()
    d.pop()
    R3=elm.Resistor().down().label('R3')
    L4 = elm.Line().left()
    elm.LoopCurrent([R1,R3,L4,V1], pad=1.25).label('$i_1$')
    elm.LoopCurrent([R1,V2,L3,R3], pad=1.25).label('$i_2$')
```
